```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
df=pd.read_excel(r'C:/Python/datasets/Year_VideoGame_ Dataset.xlsx')

print(df)
```

```python
df.head()
```

```python
plt.figure(figsize=(15,5))

plt.plot(df.Year, df.NorthAmerica_Sales, 'r.-', label = 'USA')
plt.plot(df.Year, df.EurpeanUnion_Sales, 'b.-', label = 'EU')

plt.title('USA vs EU Sales')
plt.ylabel('Sales in millions')
plt.legend()
plt.xticks(rotation=90)

plt.xticks(df.Year[::].tolist()+[2016])

plt.show()
```

```python
df.replace(to_replace ='',
               value ='0',inplace=True)
df.replace(to_replace ='NaN',
               value ='0',inplace=True)
```

```python
df['NorthAmerica_Sales'] = df['NorthAmerica_Sales'].fillna(0)
df['EurpeanUnion_Sales'] = df['EurpeanUnion_Sales'].fillna(0)
df.head()
```

```python
plt.figure(figsize=(15,8))

plt.plot(df.Year, df.NorthAmerica_Sales, 'r.-', label = 'USA')
plt.plot(df.Year, df.EurpeanUnion_Sales, 'b.-', label = 'EU')
plt.plot(df.Year, df.Japan_Sales, 'y.-', label = 'Japan')
#plt.plot(df.Year, df.Global_Sales, 'g.-', label = 'Global')

plt.title('USA / EU / Japan Sales')
plt.ylabel('Sales in millions')
plt.legend()
plt.xticks(rotation=90)

plt.xticks(df.Year[::].tolist()+[2016])

plt.show()
```
